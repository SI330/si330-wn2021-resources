```python
def generate_usernames():
    letters = 'abcdefghijiklmnopqrstuvwxyz'
    numbers = '0123456789'
    
    for x in letters:
        for y in letters:
            for z in numbers:
                for w in numbers:
                    yield x+y+z+w
    
for x in generate_usernames():
    print(x)

```
